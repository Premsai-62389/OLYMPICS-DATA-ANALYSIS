```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oly=pd.read_csv("olympics.csv")
olymp=pd.DataFrame(oly)
olymp
```

```python
olymp.rename(columns={"code":"Codes","name":"Full Name","name_short":"Nick Name","function":"Event","gender":"Gender","name_tv":"Broadcasting","country_code":"Country Code","country":"Country","nationality":"Nationality","nationality_code":"Nationality code","height":"Height","weight":"Weight","birth_date":"Birth Date","disciplines":"Disciplines"},inplace=True)
```

```python
olymp.describe()
```

```python
# Dropping a particular columns
olymp.drop(["nationality_full"],axis=1,inplace=True)
olymp.drop(["country_full"],axis=1,inplace=True)
```

```python
#Exploratary data analysis
#Univariate analysis
#Plotting distribution of 'Country' of olympics
plt.figure(figsize=(20, 30)) 
sns.histplot(olymp['Country'],color="green") 
plt.title('Country Distribution of olympics') 
plt.legend(title="count",labels=["Country"])
plt.show()  
```

```python
#Box Plotting of 'codes' for olympics
plt.figure(figsize=(10, 6)) 
plt.boxplot(olymp['Codes']) 
plt.title('Codes Distribution of olympics') 
plt.show()  
```

```python
#Count plot of events in olympics
plt.figure(figsize=(20, 20))  
plt.subplot(1, 2, 1)  
sns.countplot(x='Event', data=olymp,color="blue")
plt.legend(labels="Event")
plt.title('Events Count')
```

```python
#Count plot of Disciplines in olympics
plt.figure(figsize=(32, 20))  
plt.subplot(1, 2, 1)  
sns.countplot(x='Disciplines', data=olymp,color="purple")
plt.legend(labels="Disciplines")
plt.title('Disciplines Count in olympics')
```
